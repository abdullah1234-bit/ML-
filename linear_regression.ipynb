```python
# prompt: Predict canada's per capita income in year 2020. There is an exercise folder here on github at same level as this notebook, download that and you will find canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020

import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

# Assuming you have downloaded the 'canada_per_capita_income.csv' file
# and it's in the same directory as your notebook.
df = pd.read_csv('/canada_per_capita_income.csv')

# Prepare the data
X = df[['year']]
y = df['per capita income (US$)']

# Create linear regression object
reg = linear_model.LinearRegression()

# Train the model
reg.fit(X, y)

# Predict per capita income for 2020
year_2020 = [[2020]]
predicted_income_2020 = reg.predict(year_2020)

print("Predicted per capita income for Canada in 2020:", predicted_income_2020[0])
```

```python
# prompt: train the given model hiring.csv using linear regression

# Assuming you have downloaded the 'hiring.csv' file
# and it's in the same directory as your notebook.
d = pd.read_csv("/hiring.csv")

# Fill NaN values in 'experience' with 'zero'
d.experience = d.experience.fillna("zero")

# Convert experience to numbers using word2number
d.experience = d.experience.apply(w2n.word_to_num)

# Calculate the median of the 'test_score(out of 10)' column
median_test_score = d['test_score(out of 10)'].median()

# Fill NaN values in 'test_score(out of 10)' with the median
d['test_score(out of 10)'] = d['test_score(out of 10)'].fillna(median_test_score)

# Define X and y for regression
X = d[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = d['salary($)']

# Create and train a linear regression model
reg = linear_model.LinearRegression()
reg.fit(X, y)

# You can now use the trained model 'reg' to predict salaries for new candidates
# For example, to predict the salary for a candidate with 2 years of experience,
# a test score of 9, and an interview score of 6:
predicted_salary = reg.predict([[0, 8, 9]])
print("Predicted salary:", predicted_salary[0])
```

#Gradient descent:

```python
import numpy as np

def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.08

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print ("m {}, b {}, cost {} iteration {}".format(m_curr,b_curr,cost, i))

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```
